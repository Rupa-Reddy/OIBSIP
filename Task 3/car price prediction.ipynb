Data cleaning and preprocessing

```python
print(car_df.isnull().sum())
```

```python
car_df['car_age'] = 2024 - car_df['Year']
car_df['car_age']
```

```python
col_to_drop=['Year','Selling_type','Owner']
car_df = car_df.drop(columns=col_to_drop)
```

```python
car_df
```

```python
#checking for outliers
plt.figure(figsize=(8, 3))
sns.boxplot(x=car_df['Selling_Price'])
plt.show()
```

```python
sorted(car_df['Selling_Price'])
```

```python
car_df = car_df[~(car_df['Selling_Price'] >=33.0 ) & (car_df['Selling_Price'] <=35.0 )]
car_df 
```

Visualizing the Data

```python
import warnings
warnings.filterwarnings("ignore")
sns.pairplot(car_df[['Selling_Price', 'Present_Price', 'Driven_kms', 'car_age']])
plt.show()
```

```python
car_df['Fuel_Type'] = car_df['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
car_df['Transmission'] = car_df['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
car_df
```

```python
from sklearn.model_selection import train_test_split
# Separating Independent variable (X) and target variable (y)
X = car_df.drop(['Selling_Price', 'Car_Name'], axis=1)
Y = car_df['Selling_Price']
# Split the data into training and testing sets
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Training a model

```python
# LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
LR_model =LinearRegression()
LR_model.fit(X_train,Y_train)

#making prediction
LR_predict=LR_model.predict(X_test)

#evaluating the model
LR_MSR=mean_squared_error(Y_test,LR_predict )
LR_R2 =r2_score(Y_test,LR_predict )
print('Mean Squared Error for Linear Regression):', LR_MSR )
print('R-squared value for Linear Regression):', LR_R2 )
```

```python
#RANDOMFOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
RFR_model =RandomForestRegressor()
RFR_model.fit(X_train,Y_train)
```

```python
#making prediction
RFR_predict=RFR_model.predict(X_test)

#Evalauate
RFR_MSE=mean_squared_error(Y_test, RFR_predict)
RFR_R2 =r2_score(Y_test, RFR_predict)
print('Mean Squared Error for RansomforestRegressor):', RFR_MSE )
print('R-squared value for RansomforestRegressor):', RFR_R2 )
```

```python
#visually comparing both the models
Models = ['Linear Regression', 'RandomForest Regressor']
Rsquared_values=[LR_R2, RFR_R2]
plt.figure(figsize=(5,4))
plt.bar(Models,Rsquared_values)
plt.title('Comparison of models')
plt.xlabel('Model')
plt.ylabel('R_squared_values')
plt.show()
```

```python
if LR_R2>RFR_R2:
    print('The model linear Regression Performs Better')
else:
    print('The model Randomforest Regressor Performs Better')
    
```
