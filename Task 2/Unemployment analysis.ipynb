Data Cleaning

```python
print(unemp_inindia_data.isnull().sum())
print("\n")
print(unemp_upto11_2020_data.isnull().sum())
```

```python
#for umemployment in in india data
unepm_in_india=unemp_inindia_data.dropna()
print("after removing null values")
print(unepm_in_india.isnull().sum())
unepm_in_india.shape
```

```python
#checking for duplicates
unepm_in_india.duplicated().sum()
unemp_upto11_2020_data.duplicated().sum()
```

```python
# statistics of data frames
unepm_in_india.describe()
```

```python
unemp_upto11_2020_data.describe()
```

```python
# Convert 'Date' column to datetime
import warnings
warnings.filterwarnings("ignore")
unepm_in_india[' Date'] = pd.to_datetime(unepm_in_india[' Date'])
unemp_upto11_2020_data[' Date'] = pd.to_datetime(unemp_upto11_2020_data[' Date'])
```

```python
#creating a month column
unepm_in_india['Month'] = unepm_in_india[' Date'].dt.month
unemp_upto11_2020_data['Month'] = unemp_upto11_2020_data[' Date'].dt.month
```

```python
unepm_in_india['Month'] = unepm_in_india[' Date'].dt.strftime('%B')
unemp_upto11_2020_data['Month'] = unemp_upto11_2020_data[' Date'].dt.strftime('%B')
```

```python
print(unepm_in_india.columns)
print("\n")
print(unemp_upto11_2020_data.columns)
```

```python
# Removing Unnecessary columns for first dataset
columns_to_remove1 = [' Frequency', 'Area']
Dataframe_1=unepm_in_india.drop(columns=columns_to_remove1)
Dataframe_1.head(3)
```

```python
# Removing Unnecessary columns for first dataset2
columns_to_remove2 = [' Frequency', 'Region.1']
Dataframe_2=unemp_upto11_2020_data.drop(columns=columns_to_remove2)
Dataframe_2.head(3)
```

DATA VISUALIZATION

```python
#unemployment rate analysis by Region

#for dataframe 1
plt.figure(figsize=(15, 8))
plt.subplot(2,1,1)
sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=Dataframe_1)
plt.title(' Avg Unemployment Rate by Region(Dataframe_1)')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')

plt.xticks(rotation=90)
plt.show()

#for dataframe 2
plt.figure(figsize=(15, 8))
plt.subplot(2,1,1)
sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=Dataframe_2)
plt.title('Avg Unemployment Rate by Region(Dataframe_2)')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')

plt.xticks(rotation=90)
plt.show()
```

```python
#unemployment rate analysis by Month
import calendar
list(calendar.month_name)[1:] 

#for dataframe 1
plt.figure(figsize=(15, 8))
plt.subplot(2,1,1)
sns.lineplot(x='Month', y=' Estimated Unemployment Rate (%)', data=Dataframe_1)
plt.title(' Avg Unemployment Rate by Month(Dataframe_1)')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')

plt.xticks(rotation=90)
plt.show()

#for dataframe 2
plt.figure(figsize=(15, 8))
plt.subplot(2,1,1)
sns.lineplot(x='Month', y=' Estimated Unemployment Rate (%)', data=Dataframe_2)
plt.title('Avg Unemployment Rate by Month(Dataframe_2)')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')

plt.xticks(rotation=90)
plt.show()
```

```python
# correlation analysis

# for dataframe 1
plt.figure(figsize=(10, 8))
plt.subplot(2,1,1)
correlation_matrix_df1 = Dataframe_1[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(correlation_matrix_df1, annot=True, cmap='cividis')
plt.title('Correlation Matrix (Dataframe_1)')
plt.xticks(rotation=0)
plt.show()

#for data frame 2
plt.figure(figsize=(10, 8))
plt.subplot(2,1,1)
correlation_matrix_df1 = Dataframe_2[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(correlation_matrix_df1, annot=True, cmap='cividis')
plt.title('Correlation Matrix (Dataframe_2)')
plt.xticks(rotation=0)
plt.show()
```

```python
Dataframe_1['YearMonth'] = Dataframe_1[' Date'].dt.strftime('%Y-%m')
```

```python
#Unemployment rate in india between 05-2019 and 06-2020
sort = sorted(Dataframe_1['YearMonth'].unique())
Dataframe_1['YearMonth'] = pd.Categorical(Dataframe_1['YearMonth'], categories=sort, ordered=True)
Dataframe_1 = Dataframe_1.sort_values(by='YearMonth')
plt.figure(figsize=(15, 5))
sns.lineplot(x='YearMonth', y=' Estimated Unemployment Rate (%)', data=Dataframe_1)
plt.title('Estimated Unemployment Rate in india(05-2019 to 06-2020)')
plt.xlabel('YearMonth')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

```python
# analysis of Estimated Unemployment and Labour Participation Rates Over Months

#dataframe 1
plt.figure(figsize=(12, 5))
plt.subplot(2,1,1)
sns.lineplot(x='Month', y=' Estimated Unemployment Rate (%)', data=Dataframe_1, marker='^',label='Unemployment Rate')
sns.lineplot(x='Month', y=' Estimated Labour Participation Rate (%)', data=Dataframe_1, marker='o', label='Labour Participation Rate')
plt.title('Estimated Unemployment and Labour Participation Rates Over Months(Dataframe_1)')
plt.xlabel('Month')
plt.ylabel('Rate (%)')
plt.legend()
plt.show()
#dataframe_2
plt.figure(figsize=(12, 5))
plt.subplot(2,1,1)
sns.lineplot(x='Month', y=' Estimated Unemployment Rate (%)', data=Dataframe_2, marker='^',label='Unemployment Rate')
sns.lineplot(x='Month', y=' Estimated Labour Participation Rate (%)', data=Dataframe_2, marker='o', label='Labour Participation Rate')
plt.title('Estimated Unemployment and Labour Participation Rates Over Months(Dataframe_2)')
plt.xlabel('Month')
plt.ylabel('Rate (%)')
plt.legend()
plt.show()
```
